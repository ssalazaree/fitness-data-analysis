```python
#Pergunta 1 — Quem queima mais calorias?

#HHIIT lidera com 926 cal, mas a diferença para Cardio (885) é só de 41 calorias — ou seja, o tipo de treino sozinho não explica grande coisa nas calorias queimadas.


calorias = df.groupby('Workout_Type')['Calories_Burned'].mean().sort_values()

fig, ax = plt.subplots(figsize=(8, 5))
calorias.plot(kind='barh', ax=ax, color='steelblue')

ax.set_title('Calorias médias queimadas por tipo de treino', fontsize=14)
ax.set_xlabel('Calorias médias')
ax.set_ylabel('Tipo de treino')

# Adiciona os valores no fim de cada barra
for i, v in enumerate(calorias):
    ax.text(v + 5, i, f'{v:.0f}', va='center')

plt.tight_layout()
plt.show()
```

```python
#Pergunta 2 — Experientes treinam mais tempo?

#
duracao = df.groupby('Experience_Level')['Session_Duration (hours)'].mean().sort_values()

fig, ax = plt.subplots(figsize=(8, 5))
duracao.plot(kind='barh', ax=ax, color='coral')

ax.set_title('Duração média de treino por nível de experiência', fontsize=14)
ax.set_xlabel('Duração média (horas)')
ax.set_ylabel('Nível de experiência')

for i, v in enumerate(duracao):
    ax.text(v + 0.01, i, f'{v:.2f}h', va='center')

plt.tight_layout()
plt.show()
```
